#### Agregación de datos por categoria

```python
import numpy as np 
import pandas as pd 
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0, 500): 
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))


```

```python
income_data[1:10]
```

```python
height = 160 + 30 * np.random.randn(n)
weight = 55 + 25 * np.random.randn(n)
age = 20 + 20 * np.random.randn(n)
income = 18000 + 3000 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    { 
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head()
```

### Agrupacion de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for name, groups in grouped_gender: 
    print(name)
    print(groups)
    print("\n\n")
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender","Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group: 
    print(names)
    print(groups)
    print("\n\n")
```

#### Operaciones sobre datos agrupados

```python
double_group["Age"].filter(lambda x: x.sum() > 1860)
```

### Transformacion de variables

```python
zcore = lambda x: (x - x.mean()) / x.std()
```

```python
z_group = double_group.transform(zcore)
```

```python
import matplotlib.pyplot as plt
plt.hist(z_group["Age"])
plt.title("Zcore de la edad")
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
```

```python
double_group.transform(fill_na_mean)
```
