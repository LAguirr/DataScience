#### Fill Indexes
Reindexing -> bfill -> backward fill

```python
data = {'color': ['blue','green','yellow','red','white'],
        'object': ['ball','pen','pencil','paper','mug'], 
        'price': [1.2, 1.0, 0.6, 0.9, 1.5]}
frame = pd.DataFrame(data)
print(frame)
print(frame.reindex(method='bfill',columns=['color','price','new','object'])) #Gives the new column, the information of the next content
# in another words, the column in front the new and void column, fill backwards the column
```

```python
ser = pd.Series(np.arange(4.),index=['red','blue','yellow','white'] )
print(ser)
```

```python
ser.drop(['blue','white'])
```

```python
frame1 = pd.DataFrame(np.arange(16).reshape((4,4)),
                      index=['red','blue','yellow','white'],
                      columns=['ball','pen','pencil','paper'])
frame2 = pd.DataFrame(np.arange(12).reshape((4,3)),
                      index=['blue','green','white','yellow'],
                      columns=['mug','pen','ball'])

print(frame1 + frame2)
```

```python
print(frame1.add(frame2))
```

```python
print(frame1/frame2)
```
