### Importar el primer fichero
### Hacemos el bucle para ir recorriendo todos y cada uno de los ficheros. 
* Importante tener una consistencia en el nombre de los ficheros
* Importamos los ficheros uno a uno
* Cada uno de ellos debe apenizarse (a√±adirse al final) del primer fichero que ya habiamos cargado

### Repetimos el bucle hasta que no queden ficheros

```python
filepath = "./distributed-data/"
data = pd.read_csv(f"{filepath}/001.csv")

final_length = len(data)

for i in range(2, 333):
    if(i < 10):
        filename = f"00{i}.csv"
    elif(i < 100):
        filename = f"0{i}.csv"
    else:
        filename = f"{i}.csv"
    file = filepath + filename

    temp_data = pd.read_csv(file)
    final_length += len(temp_data)

    data = pd.concat([data, temp_data], axis=0)
```

### Inner Join
* Devuelve un data frame con las filas que tienen valor, tanto en el primer como en el segundo data frame que estamos uniendo
* El numero de filas sera igual al numero de filas **comunes** que tengas ambos data sets
    * Data set A tiene 60 filas
    * Data set B tiene 50 filas
    * Ambos comparten 30 filas
    * Entonces A Inner Join B tendr√° 30 filas
* En t√©rminos de teor√≠a de conjuntos, se trata de la intersecci√≥n de los dos conjuntos 


```python
Image(filename='./inner-join.png')
```

```python
# data_main contiene toda la info 
# data_country_dlt le falta la info de los 7 atletas
merged_inner = pd.merge(left=data_main, right=data_country_dlt,
                        how = "inner", left_on="Athlete", right_on="Athlete")
```

```python
len(merged_inner)
```

### Left Join
* Devuelve un data frame con las filas que tuvieran valor en el dataset de la izquierda, sin importar si tienen correspondencia en el de la derecha o no. 
* Las filas del data frame final que no correspondan a ninguna fila del data frame derecho, tendr√°n NAs en las columnas del data frame derecho.
* El numero de filas ser√° iguala al numero de filas del data frame izquierdo. 
    * Data Set A tiene 60 filas
    * Data set B tiene 50 filas
    * Entonces A Left Join B tendr√° 60 filas
* En terminos de teoria de conjuntos, se trata del propio data set de la izquierda quien, ademas tiene la intersecci√≥n en su interior.


```python
Image(filename='./left-join.png')
```

```python
merged_left = pd.merge(left = data_main, right = data_country_dlt,
                       how = "left", left_on = "Athlete", right_on = "Athlete")
```

```python
merged_left.head()
```

## Right Join
* Devuelve un data frame con las filas que tuvieran valor en el dataset de la derecha, sin importar si tienen correspondencia en el de la izquierda o no.
* El numero de filas ser√° igual al numero de filas del data frame derecho.
    * Data set A tiene 60 filas
    * Data set B tiene 50 filas
    * Entonces A Right Join B tendr√° 50 filas
* En terminos de teoria de conjuntos, se trata del propio data set de la derecha quien, adem√°s tiene la interseccion en su interior.

```python
Image(filename='./right-join.png')
```

### Outer Join - Full Join
* Devuelve un data frame con todas las filas de ambos, remplazando las ausencias de uno o de otro con NAs en la regi√≥n espec√≠fica.
* Las filas del data frame final que no correspondan a ninguna fila del data grame derecho (o izquierdo), tendr√°n NAs en las columnas del data grame derecho (o izquierdo).
* El n√∫mero de filas ser√° igual al m√°ximo n√∫mero de filas de ambos data frames 
    * Data Set A tiene 60 filas.
    * Data Set B tiene 50 filas. 
    * Ambos comparten 30 filas
    * Entonces A Outer Join B tendr√° 60 + 50 - 30 = 80 filas
* En terminos de teor√≠a de conjuntos, se trata de la uni√≥n de conjuntos.
