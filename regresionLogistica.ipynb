#### Ratio de posibilidades
* Cociente entre los casos de exito sobre los casos de fracaso en el suceso estudiado y para cada grupo

```python
display(Math(r'P_m = \ probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ un\ hombre'))
display(Math(r'P_f =  \ probabilidad\ de\ hacer\ compra\ sabiendo\ que\ es\ una\ mujer'))

display(Math(r'odds\in[0,+\infty]'))

display(Math(r'odds_{prchase,male} = \frac{P_m}{1-P_m} = \frac{N_{p,m}}{N_{\bar p,m}}'))
display(Math(r'odds_{purchase, female} = \frac{P,F}{1-P_F} = \frac{N_{p,f}}{N_{\bar p,f}}'))
```

```python
pm = 121/246
pf = 159/265
odds_m = pm/(1-pm) #121/125
odds_f = pf/(1-pf) #159/106
```

```python
odds_m
```

```python
odds_f
```

* Si el ratio es superior a 1, es mas probable el exito que el fracaso. Cuanto mayor es el ratio, mas probabilidad de éxito, en nuestro suceso. 
* Si el ratio es exactamente igual a 1, exito y fracaso son equiprobables (p=0.5)
* Si el ratio es menor que 1, el fracaso es mas probable que el exito. Cuanto menor es el ratio, menor es la probabilidad de exito del proceso. 


```python
display(Math(r'odds_{ratio} = \frac{odds_{purchase, male}}{odds_{o}}'))
```

```python
odds_r = odds_m/odds_f
```

```python
odds_r 
```

```python
1/odds_r # odds_f / odds_m
```

#### La regresión logística desde la regresion lineal

```python
display(Math(r'y  = \alpha + \beta \cdot x'))
display(Math(r'(x,y)\in[-\infty, +\infty]^2'))
```

```python
display(Math(r'y\in\{0,1\}'))
display(Math(r'p\in [0,1]'))
display(Math(r'P = \alpha + \beta\cdot X'))
```

P es la probabilidad condicionada de exito o de fracaso condicionada a la presencia de la variable X

```python
display(Math(r'\frac{P}{1-P} = \alpha + \beta\cdot X\in [0, +\infty]'))
```

```python
display(Math(r' In(\frac{P}{1-P}) = \alpha + \beta\cdot X'))
```

```python
display(Math(r'\begin{cases}\frac{P}{1-P}\in[0,1]\Rightarrow ln(\frac{P}{1-P}) \in[-\infty,0]\\ \frac{P}{1-P}\in[1, +\infty]\Rightarrow ln(\frac{P}{1-P}) \in[0, \infty] \end{cases}'))
```

```python
display(Math(r' In(\frac{P}{1-P}) = \alpha + \beta\cdot X'))
display(Math(r' \frac{P}{1-P} = e^{\alpha + \beta\cdot X}'))
display(Math(r' P = \frac{e^{\alpha+\beta\cdot X}}{1+e^{\alpha+\beta\cdot X}}'))
```

* Si a+bx es muy pequeño (negativo), entonces P tiende a 0
* Si a + bx =0, P = 0.5
* Si a + bx es muy grande (positivo), entonces P tiende a 1

#### Regresion logistica multiple

```python
display(Math(r' P = \frac{1}{1+e^{-\alpha+\sum_{i=1}^n\beta_i\cdot x_i}}'))
```

```python
display(Math(r' \vec{\beta} = (\beta_1, \beta_2, \cdots,\beta_n)'))
display(Math(r' \vec{X} = (x_1, x_2, \cdots, x_n)'))
display(Math(r' P = \frac{1}{1+e^{-(\alpha+\vec{\beta_i}\cdot \vec{X})}}'))
```
