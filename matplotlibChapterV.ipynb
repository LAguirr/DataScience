```python
x = np.array([0,6])
y = np.array([0,250])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
x = np.arange(100,200,10)
y = np.arange(400,500,10)
print(x, y)
```

```python
plt.plot(x,y, 'o-.k', ms=7, mec='r', mfc='yellow', linestyle='-.', lw=1)
```

```python
x1 = np.arange(100,200)
x2 = np.arange(1000)
y1 = np.arange(100,200)
y2 = np.arange(1000,2000)


plt.plot(x1,y1)
plt.subplot(1,2,1)


plt.plot(x2,y2)
plt.subplot(1,2,2)
plt.show()

```

```python
x = np.random.normal(0,8,2)
plt.hist(x)
```

```python
y = np.array([35, 25, 25, 15])
plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels= ["Apples", "Bananas", "Cherries", "Dates"]
myexplode= [0.9, 0.3, 0.5, 0.3]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels= ["Apples", "Bananas", "Cherries", "Dates"]
mycolors= ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels= ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()
```
