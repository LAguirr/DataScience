#### Fill Indexes
Reindexing -> bfill -> backward fill

```python
data = {'color': ['blue','green','yellow','red','white'],
        'object': ['ball','pen','pencil','paper','mug'], 
        'price': [1.2, 1.0, 0.6, 0.9, 1.5]}
frame = pd.DataFrame(data)
print(frame)
print(frame.reindex(method='bfill',columns=['color','price','new','object'])) #Gives the new column, the information of the next content
# in another words, the column in front the new and void column, fill backwards the column
```
