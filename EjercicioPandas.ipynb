#### Exercise 9
Write a pandas program to get the items of series 1 that are not present in series2.
sr1 = [1,2,3,4,5]
sr2 = [2,4,6,8,10]

```python
sr1 = [1,2,3,4,5]
sr2 = [2,4,6,8,10]
s1 = pd.Series(sr1)
s2 = pd.Series(sr2)
print(s1[~s1.isin(s2)])
```
